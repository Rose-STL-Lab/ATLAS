```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
basis = torch.load("group_50.pt")
invariantMetric = torch.load("metric_local.pt")
```

```python
G1 = basis[6].numpy()
G2 = basis[3].numpy()
G3 = basis[0].numpy()
G4 = basis[5].numpy()
G5 = basis[2].numpy()
G6 = basis[4].numpy()
G7 = basis[1].numpy()

#invariant metric (temporary number rn)
GT = invariantMetric.detach().numpy()
```
